```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pylab import mpl
mpl.rcParams['font.sans-serif'] = ['Arial Unicode MS'] # 指定中文默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
mpl.rcParams['figure.figsize'] = [20,20] # 设置图片大小尺寸大小
# mpl.rcParams['figure.dpi'] = 300 # 图像像素
# mpl.rcParams['savefig.dpi'] = 300 # 分辨率
```

```python
plt.figure()
datas = pd.Series(np.random.randn(1000))
datas.plot()
plt.plot(np.sin(datas),)
plt.ylabel('y轴')
plt.xlabel('x轴')
plt.title('my title')
plt.legend(('ss','ssss'))
plt.show()
```

```python
plt.subplot(2,2,1)
# plt.subplot(221)
plt.hist(datas,)
plt.subplot(2,2,2)
plt.hist(datas)
```

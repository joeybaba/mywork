```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(6,6)})
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,6):
        plt.plot(x,np.sin(x+i*.5)*( 7-i)*flip)
sinplot()
```

```python
# 
sns.set() # 
sinplot()
```

```python
# 5 darkgrid whitegrid dark white ticks
sns.set_style('ticks')
sinplot()
sns.despine() # 
sns.despine(offset=10) # 
sns.despine(left=True) # 
```

```python
# with
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

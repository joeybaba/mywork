```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(6,6)})
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,6):
        plt.plot(x,np.sin(x+i*.5)*( 7-i)*flip)
sinplot()
```

```python
# 使用设定好的模板
sns.set() # 默认组合
sinplot()
```

```python
# 5种风格 darkgrid whitegrid dark white ticks
sns.set_style('ticks')
sinplot()
sns.despine() # 删除顶角标线
sns.despine(offset=10) # 图像距离轴的距离
sns.despine(left=True) # 保留或删除轴线显示
```

```python
# with在当前风格下执行
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
# 布局大小风格
sns.set()
sns.set_context('paper') # paper talk
plt.figure(figsize=(8,6)) # 设置图片大小
sinplot()
```

```python
sns.set_context('notebook',font_scale=1.5,rc={'lines.linewidth' : 2.5})
sinplot()
```

调色板
- color_palette()能传入任何Matplotlib支持的颜色
- color_palette()不写参数默认颜色
- set_palette()设置所有图的颜色

```python
# 6个循环默认颜色 deep muted pastel bright dark colorblind
current_palette = sns.color_palette()
current_palette
sns.palplot(current_palette)
```

```python
# hls颜色空间
# sns.color_palette('hls',15)
sns.palplot(sns.color_palette('hls',15))
```

```python
data = np.random.normal(size=(20,8))
sns.boxplot(data=data,palette=sns.color_palette('hls',8))
```

```python
# hls.palette()函数控制颜色亮度和饱和 l 亮度 s 饱和 lightness saturation
sns.palplot(sns.hls_palette(n_colors=8,l=.7,s=.9))
```

```python
# Paired(配对)颜色空间
sns.palplot(sns.color_palette('Paired',10))
```

xkcd包含一套众包努力的针对随机RGB色的命名。产生954种可以随时通过xkcd_rgb字典中调用的命名颜色


```python
plt.plot([1,1],[0,1],sns.xkcd_rgb['pale red'],lw=1) # lw linewidth
plt.plot([0,1],[0,1],sns.xkcd_rgb['medium blue'],lw=1)
plt.plot([2,1],[0,1],sns.xkcd_rgb['denim blue'],lw=1)
```

```python
colors = ['windows blue','amber']
sns.palplot(sns.xkcd_palette(colors))
```

```python
# 由深到浅
sns.palplot(sns.color_palette('Blues_r'))
```

```python
# cubehelix_palette()立体螺旋调色板 色调线性变换
sns.palplot(sns.color_palette('cubehelix',8))
```

```python
sns.palplot(sns.cubehelix_palette(8,start=.05,rot=-.9))
```

```python
# 连续型
# light_palette() dark_palette() 调用定制连续调色板
sns.palplot(sns.light_palette('green'))
```

```python
sns.palplot(sns.dark_palette('green'))
```

```python
sns.palplot(sns.dark_palette('green',reverse=True))
```

```python
x,y = np.random.multivariate_normal([0,0],[[1,-.5],[-.5,1]],size=300).T # 看文档
pal = sns.dark_palette('green',as_cmap=True) # as_cmap颜色地图取代list
sns.kdeplot(x,y,cmap=pal)
```

```python
x,y = np.random.multivariate_normal([0,0],[[1,-.5],[-.5,1]],size=300).T # 看文档
pal = sns.dark_palette('green',n_colors=11)
sns.kdeplot(x,y,colors=pal)
```

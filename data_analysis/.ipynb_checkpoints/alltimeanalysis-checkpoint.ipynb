```python
start_time = time.perf_counter()
wb = openpyxl.load_workbook('data.xlsx',read_only=True)
ws = wb.active
newwb = openpyxl.Workbook()
newws = newwb.active
for i in ws.values:
    newws.append(i)
newwb.save('bb1.xlsx')
print(f'cost:{time.perf_counter() - start_time}')
```

```python
async def read():
    print('read')
    return openpyxl.load_workbook('data.xlsx', read_only=True).active

async def write():
    newwb = openpyxl.Workbook()
    newws = newwb.active
    ws = await read()
    for i in ws.values:
        newws.append(i)
    newwb.save('bb1.xlsx')

async def main():
    await asyncio.gather(write())

# IPython/ipykernel不支持asyncio，需要安装nest_asyncio
start_time = time.perf_counter()
asyncio.run(main())
# loop = asyncio.get_event_loop()
# loop.run_until_complete(main())
print(f'cost:{time.perf_counter() - start_time}')
```

```python
nf = newnew.rename(columns=newnew.iloc[0]).drop(index=0,axis=1)
nf.reset_index(drop=True,inplace=True)
nf.head(100)
```

```python
dellist_manager_center = []

for i,x in enumerate(nf['项目名称']):
    if x == '管理中心总计':
        dellist_manager_center.append(i)
    
dellist_project = []
for i,x in enumerate(nf['项目编码']):
    if x == '项目总计':
        dellist_project.append(i)
del_list = set(dellist_manager_center + dellist_project)
del_list
```

```python
nf_del_counter = nf.drop(index = del_list)
nf_del_counter.head(33)
```

```python

nf_del_counter['每日超过标准人数'] = nf_del_counter['每日超过标准人数'].astype(int)
nf_del_counter.pivot_table(values='每日超过标准人数',index='大区',aggfunc='mean').head()
```

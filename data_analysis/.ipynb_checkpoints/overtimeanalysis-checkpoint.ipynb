```python
filename = '7.xlsx'
sheet = pd.read_excel(filename,skiprows=2,header=None)
sheet.head(5)
```

```python
fillna_sheet = sheet.ffill()
fillna_sheet.head(5)
```

```python
dropna_sheet = fillna_sheet.drop([0,4,11],axis=1)
dropna_sheet.head(5)
```

```python
rename_col_sheet = dropna_sheet.rename(columns=dropna_sheet.iloc[0]).drop(index=0)
rename_col_sheet.head()
```

```python
reindex_sheet = rename_col_sheet.reset_index(drop=True)
reindex_sheet.head()
```

```python
reindex_sheet.iloc[:,5:11] = reindex_sheet.iloc[:,5:11].astype(int)
```

```python
reindex_sheet.dtypes
```

```python
reindex_sheet = reindex_sheet[~reindex_sheet['项目名称'].isin(['管理中心总计'])]
delcount_sheet = reindex_sheet[~reindex_sheet['项目编码'].isin(['项目总计'])]
delcount_sheet.head(50)
# delcount_sheet.to_excel('data_cleaning.xlsx')
```

```python
# TODO 时间转化
testsheet = delcount_sheet.copy()
testsheet.iloc[:, 11] = pd.to_datetime(testsheet.iloc[:, 11], format='%Y%m')

# testsheet.iloc[:,11] = testsheet.iloc[:,11].apply(lambda x:x.strftime('%Y%m'))
testsheet.iloc[:, 11].head()

delcount_sheet = testsheet
```

```python
# 处理完数据
```

```python
sheet = delcount_sheet.copy()
```

```python
sheet['sta_extra'] = sheet['实际每日月休人数'] - sheet['标准每日月休人数']
sheet.head()
```

```python
# 考勤时间范围
pd.to_datetime(sheet['考勤日期']).max(),
pd.to_datetime(sheet['考勤日期']).min(),
pd.to_datetime(sheet['考勤日期']).max()-pd.to_datetime(sheet['考勤日期']).min()+pd.to_timedelta('1d')
```

```python
center_sum_count = sheet.pivot_table(values='每日超过标准工时（小时）',index='大区',aggfunc='sum')
center_sum_count
```

```python
center_sum_count.plot.bar(figsize= [15,6]);
```

```python
frag_center_sum_count = sheet.pivot_table(values='每日超过标准工时（小时）',index='管理中心',aggfunc='sum')
frag_center_sum_count
```

```python
plotxl = frag_center_sum_count.plot.bar(figsize= [15,6]);
plotxl
```

```python
write = pd.ExcelWriter(newfilename)
```

```python
frag_center_sum_count.to_excel(write,sheet_name='1')
center_sum_count.to_excel(write,sheet_name='2')
write.save()
```

```python
wb = load_workbook(newfilename)
ws = wb.active
wb.sheetnames
```

```python
ws1 = wb['1']
ws2 = wb['2']
ws1,ws2
```

```python
ws1.max_column,ws1.max_row
```

```python
chart = BarChart()
chart.style = 18
chart.type = 'col'
chart.title = '各管理中心（阵地）超标准工时'
chart.x_axis.title = '管理中心（阵地）'
chart.y_axis.title = '超标准工时'
```

```python
cats = Reference(ws1,min_col=1,min_row=2,max_row=ws1.max_row)
data = Reference(ws1,min_col=2,min_row=2,max_row=ws1.max_row)
chart.add_data(data)
chart.set_categories(cats)
ws1.add_chart(chart,'D1')
```

```python
chart1 = BarChart()
chart1.style = 18
chart1.type = 'col'
chart1.title = '各大区超标准工时'
chart1.x_axis.title = '大区'
chart1.y_axis.title = '超标准工时'
```

```python
cats = Reference(ws2,min_col=1,min_row=2,max_row=ws1.max_row)
data = Reference(ws2,min_col=2,min_row=2,max_row=ws1.max_row)
chart1.add_data(data)
chart1.set_categories(cats)
ws2.add_chart(chart1,'D1')
```
